# Counts per Cell Line

```python
import numpy as np
FILEPATH='/Users/jasonmiller/WVU/Localization/TrainTest/TrainTest_ver43/'
FILENAME='all.lncRNA_RCI.csv'
```

```python
filename = FILEPATH+FILENAME
print(filename)
cell_line_names = list()
cell_line_dict = dict()
with open (filename, 'r') as fin:
    heading = None
    for line in fin:
        line = line.strip()
        if heading is None:
            heading = line
            cell_line_names = line.split(',')
            cell_line_names.pop(0) # remove gene_id
            print (cell_line_names)
            for cell in cell_line_names:
                cell_line_dict[cell] = []
        else:
            fields = line.split(',')
            fields.pop(0) # remove gene_id
            for i in range(len(fields)):
                cell = cell_line_names[i]
                string = fields[i]
                if string != 'nan':
                    value = float(string)
                    cell_line_dict[cell].append(value)
                    print(cell,string,value)
```

```python
print('Line\tCount\tMin\tMax\tMed\tMean\tStd')
for cell in cell_line_names:
    values = cell_line_dict[cell]
    count = len(values)
    minimum = np.min(values)
    maximum = np.max(values)
    median = np.median(values)
    average = np.mean(values)
    stdev = np.std(values)
    output = "%10s\t%5d\t%6.2f\t%6.2f\t%6.2f\t%6.2f\t%4.2f" % (cell,count,minimum,maximum,median,average,stdev)
    print(output)
```
