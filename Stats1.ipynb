# Statistical Significance

```python
from scipy.stats import ttest_ind as ttest
```

```python
def t_test(a,b,verbose=True):
    results = ttest(a,b) 
    t = results[0]
    p = results[1]
    if verbose:
        # print('t statistic', t)
        print('p value', p)
        print('p<0.05', p<0.05)
    return p
```

```python
def delist(list_of_list):
    list_of_float = [x[0] for x in list_of_list]
    return list_of_float
```
